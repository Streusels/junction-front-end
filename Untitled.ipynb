```python
import plotly.figure_factory as ff
import plotly.express as px

import pandas as pd

import json
```

```python
with open('export.geojson', 'r', encoding='utf-8') as f:
    geodata = json.load(f)['features']
```

```python
data = []
for building in geodata:
    geom = building['geometry']['coordinates']
    if type(geom[0]) == float:
        data.append({'lat': geom[0], 'lon': geom[1]})
        
    else:
        geom = geom[0]
        for point in geom:
            data.append({'lat': point[0], 'lon': point[1]})
```

```python
hel_df = pd.DataFrame(data)
hel_df
```

```python
fig = ff.create_hexbin_mapbox(
    data_frame=hel_df, lat="lon", lon="lat",
    nx_hexagon=70, opacity=0.5, labels={"color": "Point Count"},
    min_count=1,
)
fig.show()
```
